```python
min = []; max = []; mean = []; median = []
for i in range(freq_data.shape[0]):
    min.append(freq_data[i,:].min())
    max.append(freq_data[i,:].max())
    mean.append(freq_data[i,:].mean())
    median.append(np.median(freq_data[i,:]))
```

```python
fit = np.polyfit(np.arange(100), np.log10(max), 1)
print(fit)
x = np.arange(100)
y = fit[0] * x  + fit[1]
plt.plot(np.arange(100), np.log10(min))
plt.plot(np.arange(100), np.log10(max))
plt.plot(np.arange(100), np.log10(mean))
plt.plot(np.arange(100), np.log10(median))
plt.plot(x, y)
plt.show()
```

```python
temp = np.log10(np.array(freq_data)) - y.reshape([-1,1])
temp = np.clip(temp, -3, 2)
```
